```python
import pandas as pd
import csv
```

```python
school_data = "../Resources/schools_complete.csv"
student_data = "../Resources/students_complete.csv"

schools = pd.read_csv(school_data)
students = pd.read_csv(student_data)

public_school_data = pd.merge(students, schools, how="left", 
                              on=["school_name","school_name"])
```

```python
publicSchool_df= pd.DataFrame(public_school_data)
```

```python
TotalSchools = schools["school_name"].count()

TotalStudents = students["student_name"].count()

TotalBudget = schools["budget"].sum()

AvgMath = students["math_score"].mean()

AvgReading = students["reading_score"].mean()
```

```python
PassMath = publicSchool_df.loc[(students["math_score"] >= 70)]
PassMathCount = PassMath["math_score"].count()
MathPercent = (PassMathCount/TotalStudents)*100
```

```python
PassReading = publicSchool_df.loc[(students["reading_score"] >= 70)]
PassReadingCount = PassReading["reading_score"].count()
ReadingPercent = (PassReadingCount/TotalStudents)*100
```

```python
OverallPass = (MathPercent + ReadingPercent)/2
```

```python
District_Summary = {"Total Schools": TotalSchools,
                   "Total Students": TotalStudents,
                   "Total Budget": TotalBudget,
                   "Average Math Score": AvgMath,
                   "Average Reading Score": AvgReading,
                   "% Passing Math": MathPercent,
                   "% Passing Reading": ReadingPercent,
                   "Overall Passing Rate": OverallPass}

District_Summary_df = pd.DataFrame([District_Summary])
District_Summary_df = District_Summary_df[["Total Schools",
                                          "Total Students",
                                          "Total Budget",
                                          "Average Math Score",
                                          "Average Reading Score",
                                          "% Passing Math",
                                          "% Passing Reading",
                                          "Overall Passing Rate"]]
District_Summary_df
```

```python
copy_schools = schools.copy()
copy_schools_df = pd.DataFrame(copy_schools)
```

```python
del copy_schools_df["School ID"]
```

```python
copy_schools_df["Per Student Budget"] = copy_schools_df["budget"]/ copy_schools_df["size"]
```

```python
copy_students = students.copy()
copy_students_df = pd.DataFrame(copy_students)
```

```python
Average_Score_tbl = copy_students_df.groupby(["school_name"])["reading_score","math_score"].mean().reset_index()
```

```python
Avg_Score_df = pd.DataFrame(Average_Score_tbl)
```

```python
school_merge_df = pd.merge(copy_schools_df, Avg_Score_df, on="school_name")
```

```python
Summary_PassMath = copy_students_df[copy_students_df["math_score"] >= 70]
Summary_PassMathCount = Summary_PassMath.groupby(["school_name"])["math_score"].count().reset_index()
Summary_PassMathCount.rename({"math_score":"Math Count"}, axis=1, inplace=True)
```

```python
Summary_PassRead = copy_students_df[copy_students_df["reading_score"] >= 70]
Summary_PassReadCount = Summary_PassRead.groupby(["school_name"])["reading_score"].count().reset_index()
Summary_PassReadCount.rename({"reading_score":"Reading Count"}, axis=1, inplace=True)
```

```python
Summary_PassCount = pd.merge(Summary_PassMathCount, Summary_PassReadCount, on="school_name")
```

```python
Summary_merge_df = pd.merge(school_merge_df, Summary_PassCount, on="school_name")
```

```python
Summary_merge_df["% Passing Math"] = (Summary_merge_df["Math Count"]/Summary_merge_df["size"])*100
Summary_merge_df["% Passing Reading"] = (Summary_merge_df["Reading Count"]/Summary_merge_df["size"])*100
```

```python
del Summary_merge_df["Math Count"]
del Summary_merge_df["Reading Count"]
```

```python
Summary_merge_df["% Overall Passing"] = (Summary_merge_df["% Passing Math"]+Summary_merge_df["% Passing Reading"])/2
```

```python
Summary_merge_df.rename({"school_name":"School Name", 
                         "type":"School Type", 
                         "size": "Total Students",
                         "budget":"Total School Budget",
                        "reading_score": "Average Reading Score",
                         "math_score": "Average Math Score"
                        }, axis=1, inplace=True)
```

```python
Summary_merge_df = Summary_merge_df.set_index("School Name")
Summary_merge_df
```

```python
Top_Schools_df = Summary_merge_df.sort_values("% Overall Passing", ascending=False)
Top_Schools_df = Top_Schools_df.head()
Top_Schools_df
```

```python
Bottom_Schools_df = Summary_merge_df.sort_values("% Overall Passing")
Bottom_Schools_df = Bottom_Schools_df.head()
Bottom_Schools_df
```

```python
Math_by_Grade_df = pd.pivot_table(copy_students_df, values=["math_score"],
                                  index=["school_name"], columns=["grade"])
Math_by_Grade_df
```

```python
Reading_by_Grade_df = pd.pivot_table(copy_students_df, values=["reading_score"],
                                  index=["school_name"], columns=["grade"])
Reading_by_Grade_df
```

```python
Score_Spending_df = Summary_merge_df.copy()
```

```python
score_bins = [0, 585, 610, 635, 660]
score_labels= ["<$585", "$586-610", "$611-635", "$636-660"]
```

```python
Bin_Score_Spending = pd.cut(Score_Spending_df["Per Student Budget"],
                              score_bins, labels=score_labels)
Bin_Score_Spending_df= pd.DataFrame(Bin_Score_Spending)
Score_Spending_df["Spending Range (Per Student)"]= Bin_Score_Spending_df
```

```python
School_Spending= Score_Spending_df.groupby(["Spending Range (Per Student)"])["Average Reading Score", "Average Math Score", "% Passing Reading", "% Passing Math", "% Overall Passing"].mean()
School_Spending
```

```python
size_bins= [0, 1000, 2000, 5000]
size_labels= ["Small(<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
Size_Score_df = Summary_merge_df.copy()
```

```python
Bin_Size_Score = pd.cut(Size_Score_df["Total Students"],
                              size_bins, labels=size_labels)
Bin_Size_Score_df= pd.DataFrame(Bin_Size_Score)
Size_Score_df["School Size"]= Bin_Size_Score_df
```

```python
Size_Score_df= Size_Spending_df.groupby(["School Size"])["Average Reading Score", "Average Math Score", "% Passing Reading", "% Passing Math", "% Overall Passing"].mean()
Size_Score_df
```
